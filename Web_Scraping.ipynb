```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://books.toscrape.com/index.html'
base_url = 'https://books.toscrape.com/'
book_base_url = 'https://books.toscrape.com/catalogue/'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python

urls = []
for i in range(1, 51):
    urls.append(f'https://books.toscrape.com/catalogue/page-{i}.html')
    
print(urls)
```

```python
book_page_urls = []
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
 
    

    page_url = soup.find('h3').find('a')['href']
    
    book_page_urls.append(page_url)
print(book_page_urls)
    
    
```

```python
book_name = []
book_price = []
book_raiting = []
book_description = []
book_upc = []
book_availability = []
for book_page_url in book_page_urls:
    combines_url = book_base_url + book_page_url
    response = requests.get(combines_url)
    soup = BeautifulSoup(response.text, 'html.parser')
   
    
    title = soup.find('h1').text
    book_name.append(title)
    print(title)
    price = soup.find('p', class_='price_color').text
    book_price.append(price)
    print(price)
    raiting = soup.find('p', class_='star-rating')['class'][1]
    book_raiting.append(raiting)
    print(raiting)
    description = soup.find('div',class_='sub-header')
    description = description.find_next('p').text
    book_description.append(description)
    print(description)
    upc = soup.find('tr').find_next('td').text
    book_upc.append(upc)
    print(upc)
    availability = soup.find('tr').find_next('tr').find_next('tr').find_next('tr').find_next('tr').find_next('tr').find_next('td').text
    availability = availability[0:9]
    book_availability.append(availability)
    print(availability)
    print('-----------------------------------')

```

```python

import pandas as pd
book_data = {
    'Book Name': book_name,
    'Price': book_price,
    'Rating': book_raiting,
    'Description': book_description,
    'Availability': book_availability,
    'UPC': book_upc
}
book_df = pd.DataFrame(book_data)
book_df.to_csv('books.csv', index=False)
book_df
```
